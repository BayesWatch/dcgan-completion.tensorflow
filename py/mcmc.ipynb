```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.misc
```

```python
logs = []
for i in range(80):
    log = pd.read_table('../output/{:02d}/logs/hats_00.log'.format(i), sep=' ', index_col=0)
    logs.append(log)

logs[0]
```

```python
chains = [log.drop('loss',1) for log in logs]
```

```python
pd.concat(chains).describe()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2).fit(pd.concat(chains))
pca_chains = [pca.transform(chain) for chain in chains]

plt.plot(pca_chains[0][:,0], pca_chains[0][:,1])
plt.show()

plt.figure(figsize=(10, 10))
for i in range(10):
    plt.plot(pca_chains[00+i][:,0], pca_chains[00+i][:,1], linewidth=0.5, label='Male chain {}'.format(i))
plt.xlabel('1st principal component score')
plt.ylabel('2nd principal component score')
plt.legend(loc='best')
plt.savefig('mixing.pdf')
plt.show()

plt.figure(figsize=(10, 10))
for i in range(10):
    plt.plot(pca_chains[00+i][:,0], pca_chains[00+i][:,1], 'b', linewidth=0.5, label=None if i%10 else 'Male')
    plt.plot(pca_chains[10+i][:,0], pca_chains[10+i][:,1], 'r', linewidth=0.5, label=None if i%10 else 'Female')
plt.xlabel('1st principal component score')
plt.ylabel('2nd principal component score')
plt.legend(loc='best')
plt.savefig('gender.pdf')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
for i in range(10):
    plt.plot(logs[20+i][['loss']], 'r', linewidth=0.5, label=None if i%10 else 'Adam')
    plt.plot(logs[60+i][['loss']], 'c', linewidth=0.5, label=None if i%10 else 'L-BFGS-B')
    plt.plot(logs[00+i][['loss']], 'b', linewidth=0.5, label=None if i%10 else 'HMC sampling')
    plt.plot(logs[40+i][['loss']], 'g', linewidth=0.5, label=None if i%10 else 'HMC annealing')
plt.xlabel('Iterations')
plt.xlim([0, 1000])
plt.ylabel('Loss')
plt.ylim([0, 1000])
plt.legend(loc='best')
plt.savefig('loss.pdf')
plt.show()
```
