```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.misc
```

```python
logs = []
for i in range(50):
    log = pd.read_table('../output/{:02d}/logs/hats_00.log'.format(i), sep=' ', index_col=0)
    logs.append(log)

logs[0]
```

```python
chains = [log.drop('loss',1) for log in logs]
```

```python
pd.concat(chains).describe()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2).fit(pd.concat(chains))
pca_chains = [pca.transform(chain) for chain in chains]

plt.plot(pca_chains[0][:,0], pca_chains[0][:,1])
plt.show()

plt.figure(figsize=(20, 20))
for i in range(10):
    plt.plot(pca_chains[00+i][:,0], pca_chains[00+i][:,1])
plt.show()

plt.figure(figsize=(20, 20))
for i in range(10):
    plt.plot(pca_chains[00+i][:,0], pca_chains[00+i][:,1], 'b')
    plt.plot(pca_chains[10+i][:,0], pca_chains[10+i][:,1], 'r')
plt.show()

plt.figure(figsize=(20, 20))
for i in range(10):
    plt.plot(pca_chains[20+i][:,0], pca_chains[20+i][:,1], 'b')
    plt.plot(pca_chains[30+i][:,0], pca_chains[30+i][:,1], 'r')
    plt.plot(pca_chains[40+i][950:,0], pca_chains[40+i][950:,1], 'g')
plt.show()
```
